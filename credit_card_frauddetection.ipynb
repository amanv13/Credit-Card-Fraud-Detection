```python
##WE WILL BE USING VOTING REGRESSOR WITH DIFF REGRESSION ALGOS
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR 
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1],X_train,y_train,scoring='r2',cv=50)
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X_train,y_train,scoring='r2',cv=50)
print("Voting Regressor",np.round(np.mean(scores),2))
```

```python
##USING DECISION TREES IN VOTING REGRESSOR
```

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=3)
dt3 = DecisionTreeRegressor(max_depth=5)
dt4 = DecisionTreeRegressor(max_depth=7)
dt5 = DecisionTreeRegressor(max_depth=None)
```

```python
estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1],X_train,y_train,scoring='r2',cv=50)
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X_train,y_train,scoring='r2',cv=50)
print("Voting Regressor",np.round(np.mean(scores),2))
```

```python
##VOTING REGRESSOR COMBINED GAVE A BETTER SCORE THAN OTHER 5 DECISION TREE REGRESSOR
```
